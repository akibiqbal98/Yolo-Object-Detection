```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
im = cv2.imread("human.jpg")
```

```python
plt.imshow(im[:,:,::-1])
```

```python
bbox , conf =cv.detect_face(im)
```

```python
bbox
```

```python
conf
```

```python
labels = []
for i in range(len(conf)):
  labels.append('person')
```

```python
im2 = draw_bbox(im,bbox,labels,conf)
plt.imshow(im2)
```

```python
plt.imshow(im2[:,:,::-1])
```
